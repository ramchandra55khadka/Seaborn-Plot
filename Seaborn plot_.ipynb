```python
#Box Plot
```

```python
v1=sns.load_dataset("tips")
v1
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=v1)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=v1,hue="time")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=v1,hue="sex",color="g")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=v1,hue="sex",showmeans=True,palette="plasma")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x=v1["total_bill"])
plt.show()
```
